# Synthetic data

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2
from numpy.linalg import inv
%matplotlib notebook
```

# Generate synthetic data

```python
N = 500
np.random.seed(0)
X = 3 * np.random.uniform(-1, 1, N)
y = 10. + X - 2 * (X ** 2) + 0.5 * (X ** 3) + np.random.normal(0, 10, N)

fig,ax = plt.subplots()
ax.scatter(X,y, s=10)
```

# Use OLS to fit this data

```python
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm

poly = PolynomialFeatures(3)
Xols = poly.fit_transform(np.reshape(X,(N,1)))

model = sm.OLS(y, Xols)
results = model.fit()
print(results.summary())
```

```python
fig,ax = plt.subplots()
ax.scatter(X,y, s=10)
sorted_order = np.argsort(X)
ax.plot(X[sorted_order],results.fittedvalues[sorted_order],'r')
```

# Ex 3.2(b) : Computing the confidence interval

### Generate the grid ($x_0$) for plotting the confidence bounds

```python
x0 =np.linspace(min(X),max(X),100)
poly = PolynomialFeatures(3)
x0_ols = poly.fit_transform(np.reshape(x0,(len(x0),1)))
```

From equation 3.10 <br>
$\hat{\beta} - \beta \sim N(0, (\bf{X}^T\bf{X})\sigma^2)$
<br>
Sample $\hat{\beta} - \beta$ as a Multivariate-Normal with mean 0.
<br>
But since we do not know $\sigma$, we have from equation 3.11
<br>
$(N-p-1)\hat{\sigma}^2 \sim \sigma^2 \chi^{2}_{N-p-1}$
<br>
$\beta$ and $\hat{\sigma}$ are statistically independent.
<br>
So the confidence interval for parameters are actually given by eq 3.15 <br>
$ C_{\beta} = \left [ \beta | \left ( \hat{\beta} - \beta \right ) \bf{X}^T\bf{X} \left (\hat{\beta} - \beta \right ) \le \hat{\sigma}^2 {\chi^2_{p+1}}^{1-\alpha} \right ] $

Approach:<br>
Step 1: Sample $\beta$ from equation 3.10 with mean 0 and covariance matrix $(\bf{X}^T\bf{X})\hat{\sigma}^2$ <br>
Step 2:Check for condition in eq 3.15 from the samples and choose them for required confidence interval ($\alpha = .05$ for 95% confidence)


### Step 1: Sample $\beta$ from equation 3.10 with mean 0 and covariance matrix $(\bf{X}^T\bf{X})\hat{\sigma}^2$ <br>


```python
#Compure (XtX)^1
XTX = Xols.T.dot(Xols)
XTX_inv = inv(XTX)

# Compute sigma_hat^2
p = len(results.params)-1 # Exclude the intercept
sigma_hat_2 = np.sum((y - results.fittedvalues)**2)/float(N - p - 1)
```

```python
# Step 1:
from scipy.stats import multivariate_normal
Nsample = 10000
beta = multivariate_normal(mean=np.zeros(p+1),cov=XTX_inv*sigma_hat_2).rvs(Nsample)
```

### Step 2: Check for condition in eq 3.15 from the samples and choose them for required confidence interval ($\alpha = .05$ for 95% confidence)

```python
# Step 2:
selected_beta = np.zeros(Nsample,int)
for n in range(Nsample):
    temp = np.dot(np.matmul(beta[n,:],XTX),beta[n,:])
    if temp <= sigma_hat_2*chi2.ppf(.975,p+1) and temp >= sigma_hat_2*chi2.ppf(.025,p+1):
        selected_beta[n] = 1
    
beta_selected = beta[selected_beta==1,:] + results.params
```

```python
fig,ax = plt.subplots()
ax.scatter(X,y, s=10)
for beta_val in beta_selected:
    ax.plot(x0,np.matmul(x0_ols,beta_val))
   
```
